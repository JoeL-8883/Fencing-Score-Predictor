```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import numpy as np
```

### Visualise performance

```python
df = pd.read_csv('data.csv')
df = df.sort_values(by='date')
df['date'] = pd.to_datetime(df['date'])
```

Select fencers to retrieve data from

```python
fencer = 'MASSIALAS Alexander'
fencer2 = 'COOK Jaimie'
fencer3 = 'CHEUNG Ka Long'
df_fencer = df[df['fencer_name'] == fencer]
df_fencer2 = df[df['fencer_name'] == fencer2]
df_fencer3 = df[df['fencer_name'] == fencer3]
```

```python
plt.title('Performance')
plt.xlabel('Date')
plt.ylabel('Points')

# Plot scatter data
'''
plt.scatter(df_fencer['date'], df_fencer['points'], label=fencer, color='blue', alpha=0.7, s=5)
plt.scatter(df_fencer2['date'], df_fencer2['points'], label=fencer2, color='orange', alpha=0.7, s=5)
plt.scatter(df_fencer3['date'], df_fencer3['points'], label=fencer3, color='green', alpha=0.7, s=5)
'''

# Compute rolling averages
rolling_avg = df_fencer['points'].rolling(window=10).mean()
rolling_avg2 = df_fencer2['points'].rolling(window=10).mean()
rolling_avg3 = df_fencer3['points'].rolling(window=10).mean()

plt.plot(df_fencer['date'], rolling_avg, label=fencer, color='navy', linestyle='--')
plt.plot(df_fencer2['date'], rolling_avg2, label=fencer2, color='red', linestyle='--')
plt.plot(df_fencer3['date'], rolling_avg3, label=fencer3, color='green', linestyle='--')


plt.grid(True, alpha=0.3)
plt.legend()

plt.gca().xaxis.set_major_locator(mdates.MonthLocator(interval=12))
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
plt.xticks(rotation=45)
plt.show()
```
